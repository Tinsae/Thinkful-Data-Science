```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
seed = 2016
np.random.seed(seed)
y_0 = 1
alpha = -0.25
beta = 0.95
num=10
y = [y_0]
y_t_minus_1 = y_0

for i in range(num):
    y_t = alpha + ( beta * y_t_minus_1 ) + np.random.uniform(-1, 1)
    y.append(y_t)
    y_t_minus_1 = y_t
```

```python
plt.plot(y)
plt.show()
```
