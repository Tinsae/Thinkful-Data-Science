**Q3. Choose one continuous variable and one categorical variable, and plot them six different ways.**

- Region is categorical variable
- GDP, Literacy and Birthrate are categorical variables

```python
countries_x = countries[['Region','GDP ($ per capita)','Birthrate' ]]
countries_x = countries_x.loc[countries_x["Region"].isin(['ASIA (EX. NEAR EAST)','SUB-SAHARAN AFRICA', 'WESTERN EUROPE'])]
sns.pairplot(countries_x, hue="Region", diag_kind = "hist", markers=['o', 's', 'D'])
```

The graph above shows the relationship between GDP and birth rate for the three regions. The distribution of GDP and Birthrate is shown in a histogram separately and their relationship in a scatter plot. Africa's GDP is centered closer to 0, Asia is not too far from 0 as well but Western europe is in the higher ranges.

In technical terms the graph shows the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

```python
countries_x = countries[['Region','GDP ($ per capita)','Birthrate' , 'Climate', 'Literacy (%)']]
countries_x = countries_x.loc[countries_x["Region"].isin(['ASIA (EX. NEAR EAST)','SUB-SAHARAN AFRICA', 'WESTERN EUROPE'])]

print(countries_x.loc[countries_x['Region'] == 'SUB-SAHARAN AFRICA'].max())
print(countries_x.loc[countries_x['Region'] == 'WESTERN EUROPE'].max())
```

The four climate zones:

1. Dry tropical or tundra and ice, classification B and E.
2. Wet tropical, classification A.
3. Temperate humid subtropical and temperate continental, classification Cfa, Cwa, and D.
4. Dry hot summers and wet winters,

The boxen plot shows that Asia and Sub Saharan Africa don''t have visible countries in climate category four. Western Europe region is mostly comprised of category three and four climate.

```python
countries_x = countries_x.loc[countries_x["Climate"].isin([1,2,3,4])]
fig, ax = plt.subplots(figsize = (20,15))
ax = sns.boxenplot(x="Region", y = 'GDP ($ per capita)', hue='Climate',data=countries_x)
```

In the next graph, the emphasis is changed to Regions. Climate 3 includes has many developed countries in Western Europe. Sub-Saharan Africa has the most diverse climate it contains considerable amount of countries in all climates except category 4.

```python
fig, ax = plt.subplots(figsize = (10,8))
ax = sns.boxenplot(x="Climate", y = 'GDP ($ per capita)', hue='Region',data=countries_x)
```
